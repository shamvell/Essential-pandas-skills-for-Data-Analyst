```python
üîé Step 4: Filter Students with Math Score > 80
```

```python
df[df["Math_Score"] > 80]
```

```python
#Use bitwise AND (&)
df[(df["Math_Score"] > 80) & 
   (df["English_Score"] > 80) & 
   (df["Science_Score"] > 80)]
```

```python
#This means: check if all selected columns are greater than 90 for each row.

df[df[["Math_Score", "English_Score", "Science_Score"]].gt(80).all(axis=1)]
```

```python
#This means: check if any columns are greater than 90 for each row.
df[df[["Math_Score", "English_Score", "Science_Score"]].gt(90).any(axis=1)]
```

```python
üî¢ Step 5: Sort by English Score
```

```python
df.sort_values(by="English_Score", ascending=False).head()
```

```python
üìä Step 6: Group by Gender and Get Average Scores
```

```python
df.groupby("Gender")[["Math_Score", "English_Score", "Science_Score"]].mean()
```

```python
‚ûï Step 7: Create New Column - Total Score
```

```python
df["Total_Score"] = df["Math_Score"] + df["English_Score"] + df["Science_Score"]
df.head()
```

```python
‚úÖ Step 8: Create Pass/Fail Column
```

```python
df["Result"] = np.where(df["Total_Score"] >= 180, "Pass", "Fail")
df.head()
```

```python
üìà Step 9: Bar Chart - Average Scores by Gender
```

```python
df.groupby("Gender")[["Math_Score", "English_Score", "Science_Score"]].mean().plot(kind='bar')
plt.title("Average Scores by Gender")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```
